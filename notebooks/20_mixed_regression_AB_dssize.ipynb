## A-B Experiments

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cluster import KMeans
from imblearn.under_sampling import ClusterCentroids

from pipe_utils import build_pipeline, build_sampled_pipeline
```

```python
classes, counts = np.unique(ysyn, return_counts=True)
n_classes = len(classes)
max_samples_per_target = np.floor(np.min(counts) * 0.85) - 1

models = [[("BASE", build_pipeline())]]
for samples_per_class in np.logspace(
    np.log2(2), np.log2(max_samples_per_target), num=5, base=2, dtype=np.int32
):
    strgy = dict(zip(range(n_classes), [samples_per_class] * n_classes))
    sampler = ClusterCentroids(
        estimator=KMeans(n_init="auto", random_state=random_state),
        sampling_strategy=strgy,
        random_state=random_state,
    )

    estimator = KNeighborsRegressor(n_neighbors=1)

    pipeline = build_sampled_pipeline(sampler, estimator)
    models.append([(f"ClusterCentroids_{samples_per_class}", pipeline)])
```

```python
scoring = get_real_regression_scoring()

start_time = time.perf_counter()
results = conduct_ab_mixed_reg_experiment(
    Xsyn=Xsyn,
    ysyn=ysyn,
    Xreal=Xreal,
    yreal=yreal,
    estimators=models,
    n_splits=32,
    test_size=0.1,
    scoring=scoring,
    random_state=random_state,
)
end_time = time.perf_counter()
print(f"time passed: {end_time-start_time:.2f} s")
```
