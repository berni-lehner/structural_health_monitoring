## A-B Experiments

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cluster import KMeans
from imblearn.under_sampling import (
    ClusterCentroids,
    RandomUnderSampler,
    OneSidedSelection,
    CondensedNearestNeighbour,
)

from pipe_utils import build_pipeline, build_sampled_pipeline
```

```python
samples_per_class = 2
classes, counts = np.unique(y, return_counts=True)
n_classes = len(classes)

strgy = dict(zip(range(n_classes), [samples_per_class] * n_classes))
sampler_cc = ClusterCentroids(
    estimator=KMeans(n_init="auto", random_state=random_state),
    sampling_strategy=strgy,
    random_state=random_state,
)
sampler_rs = RandomUnderSampler(
    sampling_strategy=strgy,
    random_state=random_state,
)
strgy = [samples_per_class] * n_classes

models = [[(f"BASE", build_pipeline())]]
for sampler in [sampler_rs, sampler_cc]:
    estimator = KNeighborsRegressor(n_neighbors=1)
    pipeline = build_sampled_pipeline(sampler, estimator)
    models.append([(f"{sampler.__class__.__name__}", pipeline)])
```

```python
scoring = get_synth_regression_scoring()

results = conduct_ab_reg_experiment(
    X=X,
    y=y,
    estimators=models,
    n_splits=32,
    test_size=0.1,
    scoring=scoring,
    random_state=random_state,
)
```

```python
import time
import statistics
from rich.table import Table

tbl = Table(title="Execution times")
tbl.add_column("Model", justify="right", style="white", no_wrap=True)
tbl.add_column("Mean (ms)", justify="center", style="green", no_wrap=True)
tbl.add_column("Std. (ms)", justify="center", style="green", no_wrap=True)

for name, model in results.groupby("model"):
    model = model.iloc[0]["estimator"]
    times = []
    for _ in range(32):
        for sample in X:
            sample = sample[np.newaxis, ...]
            t1 = time.monotonic_ns()
            model.predict(sample)
            times.append(time.monotonic_ns() - t1)

    tbl.add_row(
        name,
        f"{statistics.mean(times) / 1000:.2f}",
        f"{statistics.stdev(times) / 1000:.2f}",
    )

tbl
```
